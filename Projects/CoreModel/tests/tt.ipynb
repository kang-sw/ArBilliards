```python
# test wang-hash
import numpy as np
x = np.arange(12231)
```

```python
def wang_hash(seed):
    seed = (seed ^ 61) ^ (seed >> 16)
    seed *= 9
    seed = seed ^ (seed >> 4)
    seed *= 0x27d4eb2d
    seed = seed ^ (seed >> 15)
    seed = seed & ((1<<32)-1)
    return seed
```

```python
y = [0]

for i in x:
    y.append(wang_hash(y[i]))

y = np.array(y)
y = y / np.float((1<<32)-1)
y = np.array((y & 1) * 1)
```

```python
from matplotlib import pyplot as plt

plt.hist(y)
```
